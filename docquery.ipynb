```python
from docquery import document, pipeline
```

```python
p = pipeline('document-question-answering')
```

```python
doc = document.load_document("/content/invoice.png")
```

```python
questions=["What is the invoice number?", "What is the invoice total?","unit price"]
```

```python
for q in questions:
  print(q, p(question=q, **doc.context))
```
